```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv(r'C:\Users\GEORGE TECHNOLOGY EN\Documents\data_jobs.csv')
```

```python
median_salary_year = df['salary_year_avg'].median()
```

```python
median_salary_hour = df['salary_hour_avg'].median()
```

```python
median_salary_hour
```

```python
df_filled = df
```

```python
df['salary_year_avg'] = df_filled['salary_year_avg'] .fillna(median_salary_year)
df['salary_hour_avg'] = df_filled['salary_hour_avg'] .fillna(median_salary_hour)
```

```python
df_unique = df_filled
```

```python
df_unique = df_unique.drop_duplicates()
print('df_unique length is:' ,len(df_filled))
print('df_unique length is:' ,len(df_unique))
print('Rows Dropped:', len(df_unique )-len(df_filled))
```

```python
df_unique = df_unique.drop_duplicates(subset=['job_title','company_name'])

print('df_unique length is:' ,len(df_filled))
print('df_unique length is:' ,len(df_unique))
print('Rows Dropped:', len(df_filled)-len(df_unique ))
```
