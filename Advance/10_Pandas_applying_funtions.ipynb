```python
df_salary = df[pd.notna(df['salary_year_avg'])].copy()


def projected_salary(salary):
    return salary * 1.03
df_salary['salary_infliated'] = df_salary['salary_year_avg'].apply(projected_salary)
df_salary[['salary_year_avg', 'salary_infliated']]
```

```python
# to turn the job_skill from a string to a list using the ast method 
import ast
ast.literal_eval(df['job_skills'][3])
```

```python
def clean_list(skill_list):
    if pd.notna(skill_list):
        return ast.literal_eval(df['job_skills'][3])
df['job_skills'] = df['job_skills'].apply(clean_list)
```

```python
def projected_salary(row):
    if 'Senior' in row['job_title_short']:
        return 1.03 * row['salary_year_avg']
    else:
        return 1.03 * row['salary_year_avg']
    

df_salary['salary_year_infliated'] = df_salary.apply(projected_salary, axis=1)

df_salary[['job_title_short', 'salary_year_avg', 'salary_year_infliated']]
```
